## Legend

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
#можно прописывать легенду, прям в построении графика через label
ax.plot(np.arange(0, 5, 0.25), '--', label='рост')
ax.plot(np.arange(0, 10, 0.5), ':s', label='вес')
#либо передать в легенде, label проигнорируется. Параметр loc говорит куда разместить легенду
ax.legend(['1', '2'], loc='upper right')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
#можно прописывать легенду, прям в построении графика через label
ax.plot(np.arange(0, 5, 0.25), '--', label='рост')
ax.plot(np.arange(0, 10, 0.5), ':s', label='вес')
#facecolor='#aaa' изменить цвет фона легенды, framealpha=0.5 - прозрачность
ax.legend(['1', '2'], facecolor='#aaa', fontsize=15, framealpha=0.5)
plt.show()
```

## Ступенчатый график
Используется для отображения изменений точных данных без промежуточных значений

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()

x = np.arange(0, 10)
#построение ступенчатого графика
ax.grid()
ax.step(x, x)
plt.show()
```

## Стековый график
Определяется на основе сумме, всех предыдущих значений

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
x = np.arange(-2, 2, 0.1)
y1 = np.array([- y**2 for y in x]) + 8
y2 = np.array([- y**2 for y in x]) + 8
y3 = np.array([- y**2 for y in x]) + 8
ax.stackplot(x, y1, y2, y3)
ax.grid()
plt.show()
```

## Stem-график
Если нужно отобразить данные в виде отдельных точек соединённых линиями с некой базовой величиной

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
x = np.arange(-np.pi, np.pi, 0.3)
ax.stem(x, np.cos(x))
ax.grid()
plt.show()
```

```python
#чтобы изменить базовый уровень, нужно прописать bottom
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
x = np.arange(-np.pi, np.pi, 0.3)
ax.stem(x, np.cos(x), '--r', '^g', bottom=0.5)
ax.grid()
plt.show()
```

## Точечный график
Используется для визуализации плотности

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
x = np.random.normal(0, 2, 500)
y = np.random.normal(0, 2, 500)

ax.scatter(x, y)
ax.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
x = np.random.normal(0, 2, 500)
y = np.random.normal(0, 2, 500)

ax.scatter(x, y, s=50, c='g', linewidths=1, marker='*')
ax.grid()
plt.show()
```

## Гистограммы
Она используется, когда требуется представить данные по определённым интервалам

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
y = np.random.normal(0, 2, 500)

ax.hist(y)
ax.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
y = np.random.normal(0, 2, 500)
#чтобы изменить интервалы надо дописать в hist интервалы, к примеру 50
ax.hist(y, 50)
ax.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()
y = np.random.normal(0, 2, 500)
#чтобы изменить направление графика, можно дописать orientation='horizontal'
ax.hist(y, 50, orientation='horizontal')
ax.grid()
plt.show()
```

## Круговые диаграммы
Используются, чтобы наглядно изобразить долю

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()

#значения
vals = [10, 40, 23, 30, 7]
labels = ['Toyota', 'BMW', 'Lexus', 'Audi', 'Lada']
ax.pie(vals, labels=labels)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()

#значения
vals = [10, 40, 23, 30, 7]
labels = ['Toyota', 'BMW', 'Lexus', 'Audi', 'Lada']
exp = (0, 0.2, 0.1, 0, 0)
#autopct='%.2f' - формат отображения долей, explode=exp - вынос долей за пределы графика, shadow=True -тень
ax.pie(vals, labels=labels, autopct='%.2f', explode=exp, shadow=True)
plt.show()
```
