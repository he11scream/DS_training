```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
def take_status(x):
    for i in x.split(' '):
        if '.' in i:
            return i
```

```python
data = pd.concat([train, test])
data.info()
```

```python
data.Name.apply(take_status).unique()
```

```python
data[data.Name.apply(take_status) == 'Master.']
```

```python
#Запомни эту конструкцию для изменения по выборке
data.loc[(data.Name.apply(take_status) == 'Master.') & (data.Age.isna()), 'Age'] = 6
```

```python
status = ['Rev.', 'Dr.', 'Major.', 'Lady.', 'Sir.', 'Mlle.', 'Col.', 'Capt.', 'Countess.', 'Jonkheer.', 'Dona.']
data.loc[data.Name.apply(take_status). in status]
```
