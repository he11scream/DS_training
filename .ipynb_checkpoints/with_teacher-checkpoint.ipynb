```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc("font", size=18)
```

```python
houses = pd.read_csv("1.4_houses.csv")
```

```python
houses.level.unique()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes()
ax.scatter(houses["dim_1"], houses["price"], s=100)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = houses[["dim_1"]]
y = houses["price"]
```

```python
reg = LinearRegression().fit(X, y)
```

```python
reg.intercept_
```

```python
[b] = reg.coef_
a = reg.intercept_
```

```python
def reg_prediction(dim1):
    return a + b * dim1
```

```python
reg_prediction(X.dim_1[0])
```

```python
reg.predict(X[0:2])[0:2]
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes()
ax.scatter(houses["dim_1"], houses["price"], s=100)
ax.plot([X.dim_1.min(), X.dim_1.max()], [reg_prediction(X.dim_1.min()), reg_prediction(X.dim_1.max())], c="red")
plt.show()
```

```python
X = houses[["dim_1", "dim_2"]]
y = houses["price"]
reg = LinearRegression().fit(X, y)
```

```python
[b1, b2] = reg.coef_
a = reg.intercept_

def reg_prediction(dim_1, dim_2):
    return a + b1 * dim_1 + b2 * dim_2
```

```python
reg_prediction(X.dim_1[0], X.dim_2[0])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
```

```python
X = houses[["dim_1", "dim_2"]]
y = houses["level"]
```

```python
cl = DecisionTreeClassifier().fit(X, y)
```

```python
print(export_text(cl))
```

```python
cl.predict(X[0:1])[0], y[0]
```
